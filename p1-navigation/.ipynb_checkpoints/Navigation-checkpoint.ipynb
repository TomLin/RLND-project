### 3. Take Random Actions in the Environment

In the next code cell, you will learn how to use the Python API to control the agent and receive feedback from the environment.

Note that **in this coding environment, you will not be able to watch the agent while it is training**, and you should set `train_mode=True` to restart the environment.

````python
env_info = env.reset(train_mode=True)[brain_name] # reset the environment
state = env_info.vector_observations[0]            # get the current state
score = 0                                          # initialize the score
while True:
    action = np.random.randint(action_size)        # select an action
    env_info = env.step(action)[brain_name]        # send the action to the environment
    next_state = env_info.vector_observations[0]   # get the next state
    reward = env_info.rewards[0]                   # get the reward
    done = env_info.local_done[0]                  # see if episode has finished
    score += reward                                # update the score
    state = next_state                             # roll over the state to next time step
    if done:                                       # exit loop if episode finished
        break
    
print("Score: {}".format(score))
````

### 5. Pre-defined Functions.
Two supplementary functions are defined.
- `train_dqn` sets up the training session's configuration, such as **number of episodes, epsilon start and decay rate** etc.
- `plot_scores`

````python
import torch
from collections import deque

# train the agent
def train_dqn(agent, memory, n_episodes=2000, max_step=1000, eps_start=1., eps_end=0.01, eps_decay=0.995,
              mode='train', pth='checkpoint.pth'):
    '''Set up training's configuration and print out episodic performance measures, such as avg scores, avg loss.
    
    Params
    ======
        agent (class object)
        memory (class attribute): agent's attribute for memory size tracking
        mode (string): 'train' or 'test', when in test mode, the agent acts in greedy policy only
        pth (string path): file name for the checkpoint    
    '''
    
    scores = []
    scores_window = deque(maxlen=100)  # last 100 scores
    loss_window = deque(maxlen=100) 
    epsilon = eps_start

    for i_episode in range(1, n_episodes+1):
        env_info = env.reset(train_mode=True)[brain_name]  # reset the environment and activate train_mode
        state = env_info.vector_observations[0]            # get the current state
        score = 0
        agent.running_loss = 0
        agent.training_cnt = 0
        for step in range(max_step):
            action = agent.act(state, mode, epsilon)
            env_info = env.step(action)[brain_name]        # send the action to the environment
            next_state = env_info.vector_observations[0]   # get the next state
            reward = env_info.rewards[0]                   # get the reward
            done = env_info.local_done[0]                  # see if episode has finished        

            agent.step(state, action, reward, next_state, done)

            score += reward
            state = next_state
            if done:
                break

        scores_window.append(score)
        scores.append(score)
        loss_window.append(agent.running_loss/(agent.training_cnt+0.0001)) # avoid zero
        epsilon = max(eps_end, eps_decay*epsilon) # decrease epsilon
        print('\rEpisode {:>5}\tAverage Score:{:>10.2f}\tMemory Size:{:>6}\tLoss:{:.8f}'.format(
            i_episode, np.mean(scores_window), len(memory), np.mean(loss_window)), end="")
        if i_episode % 100 == 0:
            print('\rEpisode {:>5}\tAverage Score:{:>10.2f}\tMemory Size:{:>6}\tLoss:{:.8f}'.format(
                i_episode, np.mean(scores_window), len(memory), np.mean(loss_window)))
    torch.save(agent.qnetwork_local.state_dict(), pth)

    return scores
````

````python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

def plot_scores(scores, rolling_window=100):
    '''Plot score and its moving average on the same chart.'''
    
    fig = plt.figure(figsize=(10,5))
    plt.plot(np.arange(len(scores)), scores, '-c', label='episode score')
    plt.title('Episodic Score')
    rolling_mean = pd.Series(scores).rolling(rolling_window).mean()
    plt.plot(np.arange(len(scores)), rolling_mean, '-y', label='rolling_mean')
    plt.ylabel('score')
    plt.xlabel('episode #')
    plt.legend()
````

### 6. Experiment Procedures
1. Here, three experiments will be implemented. In first experiemnt, agent acts based on Deep Q Network with uniformly probalistic sampling on replay memory. In second experiment, agent acts on Double DQN along with uniform sampling. In third experiment, agent acts on Double DQN with prioritized replay memory.
2. All agents share the same hyperparameters and random seed. The way, I can better track and differentiate the performances among different algorithms. 

### 7. Model Structure
1. For it's a low dimensional state space (instead of raw pixels), I use fully-connected layers only to build up the neural netowrk for state-action functional approximator. The network has two layers, each with 64 units and relu activation. The final output layer is 4 units without activation function. It is a rather simple architecture.

### 8. Hyperparameters
* Replay memory size: 50000
* Sample batch size: 32
* Gamma: 0.99 (the discount rate for future rewards)
* Learning per step: 1 (agent learns at every step)
* Update per step: 500 (copy the parameters of evaluating network to target network every 500 steps)
* Learning rate:
    1. For agent with uniform experience sampling, the learning rate is 0.00025, which refereces from [prioritized experience replay](https://arxiv.org/abs/1511.05952)
    2. For agent with prioritized experience sampling, the learning rate is 0.00025/4, for the variance of the stochastic gradient tends to be larger, I balance it out by decreasing the learning rate, again, the advice from the paper above.

### Agent 03: DDQN With Prioritized Sampling
In prioritized sampling, there are some modifications worth noted. First, the reward and TD error clipping. As suggested from the original paper, in order to futher stabalize the training, the reward and TD error are both clipped to fall within range (-1,1) via tanh function. I tried to remove the clipping limitation but found out that the training loss becomes more volatile and even increases exponentially along the time. Thus, I put back the clipping practice. In addition, there are two version of prioritization - proportional prioritization vs rank-based prioritization. In this post, proportional prioritization is conducted.

Secondly, in order to perform prioritized sampling, I took great reference from the following two source script:
1. [DQN with Prioitized Experience Replay](https://github.com/MorvanZhou/Reinforcement-learning-with-tensorflow/tree/master/contents/5.2_Prioritized_Replay_DQN)
    - It uses SumTree class to store the experience, which consists of two objects. One is the tree itself. The tree's leaves contain the importance (priority in the paper) of each sample, serving the weights for probalistic sampling.
    - The other object is the experiences container - the memory. In original script, numpy array is used to store these experiences. However, in order to be consistent to the memory we use in previous DQN agent, I modify the code to accommodate the experience namedtuple, thus more compatible to our other script. One final note, the container for these experience namedtupes is ordinary list object instead of deque list we used earlier. 
2. [PER in PyTorch](https://github.com/rlcode/per)
    - The first script is run on tensorflow. Hence, I search for another reference for implementation in PyTorch. Here is what I found. The basic structure of the script is almost the same, except in PyTorch syntax. 
    - One thing worth noted here is that the normalization of Importance-Sampling Weights in this post is computed based on each sampling batch. That says, these weights are not normalized based on max weights of all experiences but on the max weights within that current sampling batch. It is a bit like batch normalization. I adopt the practice in this experiment as well.
    
There are some additional hyperparameters only belonging to prioritized sampling. I list these parameters and their settings here.
* $\alpha$: 0.6 (It is within range [0,1], which is a trade-off between totally uniformly sampling and totally prioritized sampling.)
* $\beta$: 0.4 (adjusting factor for importance-sampling weights, which increases to 1 incrementally.)

````python
from prioritizedagent import PrioritizedAgent

agent_p_ddqn = PrioritizedAgent(state_size=37, action_size=4, seed=25, is_double_q=True)
````

````python
p_ddqn_scores = train_dqn(agent=agent_p_ddqn, memory=agent_p_ddqn.prioritized_memory,
                          n_episodes=2500, pth='./checkpoint/agent_p_ddqn_checkpoint.pth')

np.save('./data/p_ddqn_scores.npy', p_ddqn_scores)

plot_scores(p_ddqn_scores)
````

### 10. Conclusion
1. As expected, the DDQN with prioritized experience reply indeed outperforms other two experiments. On the other hand, if we look at the performance of DQN and DDQN, the DDQN does better than DQN, but not significantly in this case.
2. The agent successfully learns to receive episodic scores above the target +13 scores. Specifically,
    - DQN agent reaches the target at around episode 1100.
    - DDQN agent reaches the target at around episode 800.
    - DDQN with prioritized sampling reaches the target at around episode 500.
2. We can tell there is a ceiling for the average score the agent can get. It reaches plateau limit at around +15 to +17. I guess maybe it's because the neural network for state-action is a bit simple, that it reaches its maximal predicting power. Possibly if I set up a more complicated neural network, the performance would improve.
3. Let's compare the untrained and trained agent. The untrained agent who acts randomly receives scores around 0 to 1. However, the trained agent can receive scores around +20. That's a big improvement.
4. Lastly, during the previous trials, I set up the same learning rate for ordinary DQN and prioritized one. I did observe sometimes the prioritized DQN's perforamnce oscillates and gets stuck in between. I think it re-affirms that we should further decrease the learning rate in prioritized experience replay scenario.
